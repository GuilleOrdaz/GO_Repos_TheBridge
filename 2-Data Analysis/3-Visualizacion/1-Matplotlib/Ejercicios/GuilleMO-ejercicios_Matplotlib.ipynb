22. Contornos de 3D a 2D. Escribe una funci칩n que recibiendo x,y devuelva (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

Tablero: Tanto x como y van de -3 a 3 y usaremos 256 puntos.
Usa contourf, con 8 niveles (cortes), una transparencia de 0.75 y un color map de tipo 'jet'. 
Bonus: pinta las l칤neas de los contornos tambi칠n de negro

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
```
