1) Write a Python program to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).

```python
# Your code here

def checker(text):
    condicion= re.compile(r'[\w]')
    text = condicion.search(text)
    return   bool(text)

print(checker("GUIL$E97"))
print(checker('$$%/@'))
```

3) Write a Python program that matches a string that has an a followed by one or more b's

```python
# Your code here

def one_bs(texto1):
    pattern = 'b+'
    if re.search(pattern, texto1):
        return ("todo OK")
    else: 
        return("Te peinas")

print(one_bs("hola"))
print(one_bs("Burroconb"))

#la condicion es que al menos aparezca 1 vez la b. si no aparece "te peinas"
```

7) Write a Python program to find sequences of lowercase letters joined with a underscore.

```python
# Your code here
def one_bs(texto1):
    pattern = '[^A-Z.]'
    if re.search(pattern, texto1):
        return ("todo OK")
    else: 
        return("Te peinas")

print(one_bs("HOLA"))
print(one_bs("burro_con_b"))

#duda ¿por qué al marcar como contrario a las mayusculas me lo da por buena? 
```

8) Write a Python program to find the sequences of one upper case letter followed by lower case letters.

```python
# Your code here
def one_bs(texto1):
    pattern = '[A-Z]{1}[a-z]' 
    if re.search(pattern, texto1):
        return ("todo OK")
    else: 
        return("Te peinas")

print(one_bs("Cesar de la HOZ"))
print(one_bs("Burroconb"))
```

9) Write a Python program that matches a string that has an 'a' followed by anything, ending in 'b'.

```python
# Your code here

def one_bs(texto1):
    pattern = 'a.*b$' 
    if re.search(pattern, texto1):
        return ("todo OK")
    else: 
        return("Te peinas")

print(one_bs("laCesar de la HOZb"))
print(one_bs("javi"))

#la condicion es que tenga una a y termine en b
```
